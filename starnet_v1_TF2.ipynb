```python
import tensorflow as tf
from PIL import Image as img
import logging
tf.get_logger().setLevel(logging.ERROR)
from starnet_v1_TF2 import StarNet
```

```python
tf.version.VERSION
```

```python
starnet = StarNet(mode = 'RGB', window_size = 512, train_folder = './train/', lr = 1e-4)
```

```python
starnet.load_training_dataset()
```

```python
starnet.load_model('./weights', './history')
```

```python
starnet.train(1, plot_progress = True, plot_interval = 500)
```

```python
starnet.save_model('./weights', './history')
```

```python
starnet.plot_history(last = 100000)
```

<h4>Image transformation example. Works with 8bit and 16bit TIFF files</h4>

```python
in_name = "rgb_test5_16.tif"
out_name = "rgb_test5_16_starless.tif"
starnet.transform(in_name, out_name)
```
